# seaborn


## Load Dataset

```python
import pandas as pd
```

```python
url = 'https://bit.ly/dsa-04-health'

data = pd.read_csv(url)

print(data.shape)
data.head()
```

## Explore

### 디폴트 옵션 

```python
import matplotlib
import seaborn as sns
from IPython.display import set_matplotlib_formats

#windows
#matplotlib.rc('font', family = 'NanumGothic')
matplotlib.rc('font', family= 'Malgun Gothic')

#mac
#matplotlib.rc('font', family = 'AppleGothic')

#화질 선명하게(레티나 디스플레이급으로)
set_matplotlib_formats('retina')

#Minus
matplotlib.rc('axes', unicode_minus = False)
```

```python
sns.countplot(data=data, x='성별코드')
```

## 사이즈 조절하기

```python
import matplotlib.pyplot as plt

plt.figure(figsize = (18, 3))

sns.countplot(data=data, x = '시도코드')
```

### 여러개의 시각화를 한 번에 띄우기

```python
figure, (ax1, ax2) = plt.subplots(nrows = 1, ncols = 2)

figure.set_size_inches(18, 3)

sns.countplot(data=data, x = '성별코드', ax = ax1)
sns.countplot(data=data, x = '시도코드', ax = ax2)
```

```python
# figure, ((ax1)) = plt.subplots(nrows = 1, ncols = 1)
# figure, ((ax1, ax2, ax3)) = plt.subplots(nrows = 1, ncols = 3)

figure, ((ax1, ax2, ax3), (ax4, ax5, ax6)) = plt.subplots(nrows = 2, ncols = 3)
```

```python
figure, axes = plt.subplots(nrows = 3, ncols = 10)
#직접적으로 연관은 없지만 간접적으로 연관이 있어서 상관관계를 보기위해 하나의 시각화로 묶어서 띄울때 (예. 시 정보, 분정보, 초정보 )
```

## countplot

```python
plt.figure(figsize = (4, 9))
sns.countplot(data=data, y='시도코드', hue = '성별코드')

#hue는 컬럼을 쪼개서 하나의 그래프를 더생성
#x,y는 축
```

```python
sns.distplot(data['허리둘레'])

#distplot은 kdeplot을 메인으로 신뢰, bar는 histogram(sub data)
```

```python
sns.distplot(data['허리둘레'], hist = False)

```

```python
sns.distplot(data['허리둘레'], kde = False)
```

```python
data['흡연상태']
```

```python
one = data[data['흡연상태'] == 1]
two = data[data['흡연상태'] == 2]
three = data[data['흡연상태'] == 3]

sns.distplot(one['허리둘레'], hist = False, label = '흡연 안 함')
sns.distplot(two['허리둘레'], hist = False, label = '흡연 했지만 그만 둠')
sns.distplot(three['허리둘레'], hist = False, label = '흡연 하고 있음')
```

## barplot

```python
plt.figure(figsize = (18, 3))
sns.barplot(data=data, x="연령대코드(5세단위)", y = '수축기혈압')
```

```python
import numpy as np

plt.figure(figsize = (18, 3))
sns.barplot(data=data, x="연령대코드(5세단위)", y = '수축기혈압', estimator = np.sum) #np.mean(평균)도 가능 sum(은 시간당 누적매출)
```

```python

plt.figure(figsize = (4, 9))
sns.barplot(data=data, y="연령대코드(5세단위)", x = '수축기혈압', hue = '성별코드', orient = 'h') 
```

## boxplot

```python
plt.figure(figsize = (18, 3))
sns.boxplot(data=data, x = '연령대코드(5세단위)', y = '수축기혈압')
```

```python
#네모박스 가운데가 평균(mean), 가운데의 선이 median
#데이터가 쏠려있거나 아웃라이어가 많으면 메디안과 평균이 안맞는 경우가 있음(연봉)
#박스 외곽의 선은 최대값, 최소값 (바깥에 나간 데이터는 다이아몬드로 표시 < 아웃라이어)
```

```python
plt.figure(figsize = (18, 3))
sns.boxplot(data=data, x = '연령대코드(5세단위)', y = '수축기혈압', hue = '성별코드')
```

## violinplot


```python
plt.figure(figsize = (18, 3))

sns.violinplot(data=data, x = '연령대코드(5세단위)', y = '수축기혈압')
```

```python
#데이터의 전반적인 분포를 볼 수 있음. 
#min, max를 넘어서서 전반적인 분포를 보고싶을 때 가끔 씀
```

```python
plt.figure(figsize = (18, 3))

sns.violinplot(data=data, x = '연령대코드(5세단위)', y = '수축기혈압', hue = '성별코드')
```

```python
plt.figure(figsize = (18, 3))

sns.violinplot(data=data, x = '연령대코드(5세단위)', y = '수축기혈압', hue = '성별코드', split = True)
```

```python
#split = True성별코드를 나눠서보는게 아니라 왼쪽은 여성, 오른쪽은 남성일때의 분포
```

```python
#주로 박스플롯을 쓰지만 전반적인 분포가 중요한 상황일때는 바이올린플롯을 쓴다.
```

## pointplot

```python
plt.figure(figsize = (18, 3))

sns.pointplot(data=data, x = '연령대코드(5세단위)', y = '수축기혈압', hue = '성별코드', linestyles = [':', '--'])
```

```python
#왼쪽에서 오른쪽으로 선을 그려주는 개념.
#x축이 서로간에 연관성이 없으면 barplot, (ex. 시도코드)
#x축이 서로 연관성이 있고, 특히 시계열에 따른 연관이있을때 씀(시간데이터))
#seaborn linestyles => 구글링. 
```
