## pivot table ->엑셀보다 대용량 데이터에 최적화

```python
pd.pivot_table(order,index ="state", values ="price", aggfunc = ['mean', 'sum'])
#NaN은 기본적으로 배제됨
#mean은 평균, sum은 누적
```

```python
#넘파이는 커스터마이즈드된 함수/수학연산 편하게 해주는 패키지 아래 형태로도 표현 가능
import numpy as np

pd.pivot_table(order, index = 'state', values='price', aggfunc = [np.mean, np.sum])
```

```python
pd.pivot_table(order, index = ['date', 'state'], values = 'price')
```

## Group by

```python
order['price'].mean() #평균금액
```

```python
order.groupby('state')['price'].mean()

#스테이트가 캔슬드일때랑 컨펌드일때 mean값 따로 계산
```

```python
order.groupby(['date', 'state'])['price'].sum()
#피봇테이블과 그룹바이는 거의 비슷함.
#통계결과를 눈으로 확인하고 싶으면 피봇테이블, 그다음 단계가 있으면 그룹바이
#피봇테이블이 예쁘게 출력되는편
```

```python
%timeit order.groupby(['date', 'state'])['price'].sum()
```

```python
%timeit pd.pivot_table(order, index = ['date', 'state'], values = 'price')

# 그룹바이가 피벗테이블보다 열배정도 빠르다. 
```

```python
table = pd.pivot_table(order, index = ['date', 'state'], values = 'price')

table.sort_values(by = 'price', ascending = False ).head()
#피봇테이블도 메소드체인이 가능하다.
```

## apply

```python
def is_vip(price):
    if price >= 500 :
        return "VIP"
    else:
        return "Non-VIP" 
    
order['price'].apply(is_vip)
```

```python
def is_vip(row):
    if row['state'] == 'confirmed' and row['price'] >= 500:
        return 'VIP'
    else: 
        return 'Non-VIP'

order.apply(is_vip, axis = 'columns')
```

## 여러개의 데이터를 하나로 합치기

```python
url = "https://bit.ly/dsa-02-order-201701"
order201701 = pd.read_csv(url, index_col = 'id')
order201701
```

```python
url = "https://bit.ly/dsa-02-order-201702"
order201702 = pd.read_csv(url, index_col = 'id')
order201702
#데이터는 쪼개주는게 좋은거다. 잘못되면 잘못된부분만 다시 다운받을수있으니
```

```python
#concat == concatenate. 데이터 합치는 기능
order = pd.concat([order201701, order201702])
order
```

```python
url = "http://bit.ly/dsa-02-order-how"

how = pd.read_csv(url, index_col = 'id')
how
```

```python
pd.concat([order, how], axis = "columns")
```

```python
pd.merge(order, user)
#판다스가 직접 찾는 기능도 있으나, 컴퓨터가 찾는거니까 좀 위험하긴 함.
```

```python
pd.merge(order, product, left_on = "product_id", right_on = "id")
#키는 가능한 명시적으로 지정을 해야하고, 키가 아닌데 동일한 컬럼이 합쳐질 경우 왼쪽은 x, 오른쪽은 y라는 단어가 추가됨.
```

```python
pd.merge(order,user, on = 'user_id', how = "inner") 
#inner merge 는 왼쪽과 오른쪽에 모두있는것만 합친다. 둘중 하나라도 없으면 버린다. 교집합
#outer merge는 둘중 하나만 있으면 살린다. 없는 정보는 NaN. 합집합
#left : 왼쪽은 살리고, 오른쪽은 버린다.
#right: 오른쪽은 살리고, 왼쪽을 버린다.

#추천: 분석의 메인이되는 주제가 해당하는 메인데이터를 왼쪽에놓고 left merge하는 것을 추천. 보통 결과가 잘 나옴.
#잘안나오면 right, inner, outer를 고려
```
