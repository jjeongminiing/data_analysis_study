```python
def is_vip(amount):
    if amount >= 500:
        return "VIP"
    else: 
        return "None-VIP"

order["status"] = order["amount"].apply(is_vip)

#파이썬의 for문 역할
#같은기능을 판다스로 구현하는 방법을 모르는데, 파이썬으로 할 수 있으면 파이썬 문법으로 함수만들고 apply 사용

order
```

```python
def is_vip(row):
    amount = row["amount"]
    result = row["result"]
    
    if amount >= 500 and result == "confirm" :
        return "VIP"
    
    else: 
        return "None-VIP"
    
    return result

order["status"] = order.apply(is_vip, axis = "columns")

order
```
