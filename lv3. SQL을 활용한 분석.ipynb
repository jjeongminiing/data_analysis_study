```python
import sqlite3
import pandas as pd
```

# Load Dataset

```python
connection = sqlite3.connect("e-commerce.sqlite")
#경로, 파일명 정확히 입력
connection
```

```python
query = "SELECT id, date, price FROM 'orders'"

order = pd.read_sql(query, connection, index_col = "id")

order
```

```python
query = "SELECT * FROM 'orders' ORDER BY date DESC"
#날짜순 정렬
order = pd.read_sql(query, connection, index_col = "id")

order
```

```python
query = """
    SELECT *
        FROM 'orders'
        ORDER BY date DESC
        LIMIT 3
"""

order = pd.read_sql(query, connection, index_col = "id")

order
```

```python
query = "SELECT * FROM 'orders' WHERE price >= 500"
#가격이 500이상인것만 찾아줘

order = pd.read_sql(query, connection, index_col = "id")

order
```

```python
query = "SELECT * FROM 'orders' WHERE state <> 'confirmed'"
#<>가 !=라는뜻

order = pd.read_sql(query, connection, index_col = "id")

order
```

```python
query = "SELECT * FROM 'orders' WHERE address IN ('Seoul', 'Incheon')"
#adress가 seoul이나 incheon둘중 하나에 포함되면 가져와

order = pd.read_sql(query, connection, index_col = "id")

order
```

```python
query = """
    SELECT * 
        FROM 'orders' 
    WHERE price>= 500 
    AND state = 'confirmed'
    ORDER BY price DESC
    LIMIT 3
"""

#adress가 seoul이나 incheon둘중 하나에 포함되면 가져와

order = pd.read_sql(query, connection, index_col = "id")

order
```

```python
query = "SELECT * FROM 'orders' WHERE address LIke '%D%'"
#D라는 문자가 포함된거 찾아줘
order = pd.read_sql(query, connection, index_col = "id")

order
```

# 여러개의 데이터를 하나로 합치기

```python
query = "SELECT * FROM 'users'"

users = pd.read_sql(query, connection, index_col = "id")

users
```

```python
query = "SELECT * FROM 'orders'"

order = pd.read_sql(query, connection, index_col = "id")

order
```

```python
query = """
    SELECT * 
        FROM 'orders'
    LEFT JOIN 'users' ON 'orders'.user_id = 'users'.id
"""
#pd의 merge기능
pd.read_sql(query, connection)

```

```python
query = """
    SELECT 'orders'.id, 'orders'.date, 'orders'.price, 'users'.name, 'users'.gender
        FROM 'orders'
    LEFT JOIN 'users' ON 'orders'.user_id = 'users'.id
"""
#inner, outer, right 가능
#별표보다는 필요한 컬럼 찝어주는걸 추천
pd.read_sql(query, connection)
```
